```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Netflix_Data.csv')
```

```python
print(df.shape)
print(df.columns)
df.head(10)
```

```python
df.tail(5)
```

Checking for missing values

```python
df.isnull().sum()
```

```python
df_2 = df.copy()
df_2['director'].fillna('Unknown', inplace=True)
df_2['cast'].fillna('Unknown', inplace=True)
df_2['country'].fillna('Unknown', inplace=True)
```

```python
df_2[df_2['rating'].isnull()]
```

```python
#Dropping 4 rows with missing ratings as they are insignificant compared to the dataset size

df_2.dropna(subset=['rating', 'date_added', 'duration'], inplace=True)
```

```python
df_2.isnull().sum().sum()
```

```python
# Find the rows that will become NaT
problematic_dates = df_2[~df_2['date_added'].str.match(r'^[A-Za-z]+ \d{1,2}, \d{4}$', na=False)]
print(f"Number of problematic dates: {len(problematic_dates)}")
print("Problematic dates:")
print(problematic_dates['date_added'].unique())
```

```python
# Remove leading/trailing whitespace first
df_2['date_added'] = df_2['date_added'].str.strip()

# Now convert to datetime
df_2['date_added'] = pd.to_datetime(df_2['date_added'], format='%B %d, %Y')

# Verify no NaT values
print(f"NaT values after cleaning: {df_2['date_added'].isna().sum()}")
print(df_2['date_added'].dtypes)
```

```python
df_2['month_added'] = df_2['date_added'].dt.month_name()
df_2['year_added'] = df_2['date_added'].dt.year
df_2.head(3)
```

```python
df_2.info()
```

### Objective:
- To combine different aspects of our analysis to uncover more complex relationships and generate actionable insights.

```python
data =  df_3[df_3['listed_in'].isin(['Dramas', 'Comedies', 'Documentaries'])]
data2 = data.groupby('year_added')['listed_in'].value_counts().reset_index(name='count')
sns.lineplot(data=data2, x='year_added', y='count', hue='listed_in')
plt.show()
```

- Documentaries surged massively after 2015, peaking around 2017–2019 — way ahead of other genres.
- Dramas and Comedies grew steadily post-2016, with Dramas slightly leading overall.
- All genres dipped after 2019, likely due to production slowdowns or fewer releases.
- Documentaries dominated overall, showing strong viewer and platform focus on factual content.

```python
# Directly group by first country without creating a new column
country_counts = (df_2['country']
                 .str.split(',')
                 .str[0]
                 .str.strip()
                 .value_counts()
                 .reset_index())
country_counts.columns = ['country', 'count']
```

```python
# Get top 5 countries
top_5_countries = country_counts.head(5)['country'].tolist()

# Filter dataframe for top 5 countries and extract first country
df_top5 = df_2[df_2['country'].str.split(',').str[0].str.strip().isin(top_5_countries)].copy()
df_top5['first_country'] = df_top5['country'].str.split(',').str[0].str.strip()

# Group by year and country to get counts
country_year_counts = df_top5.groupby(['release_year', 'first_country']).size().reset_index(name='count')

# Pivot for area chart
pivot_data = country_year_counts.pivot(index='release_year', columns='first_country', values='count').fillna(0)
```

```python
plt.figure(figsize=(14, 8))
for country in top_5_countries:
    country_data = country_year_counts[country_year_counts['first_country'] == country]
    plt.plot(country_data['release_year'], country_data['count'], label=country, marker='o', linewidth=2)

plt.title('Netflix Content Growth by Top 5 Countries Over Time')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

- U.S. dominates Netflix content, showing a massive spike around 2015–2020 — way higher than any other country.
- India and the U.K. grew sharply after 2015, signaling Netflix’s global expansion phase.
- Canada and ‘Unknown’ sources show moderate growth but stay behind the top three.
- All countries dip post-2020, likely due to content saturation or pandemic-related production halts.
